# Example of usage of <i>[Scats.jl](https://github.com/Paveloom/Scats.jl) package</i>

Generating time series, visualizing them

Loading packages:

```julia
using Scats  # Spectral correlation analysis of time series
using PyPlot # Matplotlib (Python) wrapper for Julia
```

Setting plots parameters:

```julia
# Get access to rcParams
rcP = PyPlot.PyDict(PyPlot.matplotlib."rcParams")

# Set DPI
rcP["savefig.dpi"] = 300
rcP["figure.dpi"] = 300

# Activate TeX support
rcP["text.usetex"] = true

# Activate additional languages support: Russian
if Sys.iswindows()
    rcP["text.latex.preamble"] = [raw"\usepackage[main=russian,english]{babel}"]
else
    rcP["text.latex.preamble"] = [raw"\usepackage[main=russian,english]{babel}", raw"\usepackage{cmlgc}"]
end

# Set fontset for mathtext
rcP["mathtext.fontset"] = "cm"

# Set font sizes
rcP["font.size"] = 18
rcP["legend.fontsize"] = 12;
```

Generating time series:

```julia
s = Scats.api() # Create an instance of API

s.read_gen!("files/gen") # Read generator parameters

s.gen!() # Generate time series

s.write_input("files/input") # Write input data in a file
```

Visualizing time series:

```julia
# Create a plot
plot(s.input.t, s.input.x, color="#425378")

# Add a title
title(raw"\textrm{Time series}")

# Add labels
xlabel(raw"\textrm{Time}")
ylabel(raw"\textrm{Value}")

# Save the figure
savefig("figures/input.pdf", bbox_inches="tight")
```
